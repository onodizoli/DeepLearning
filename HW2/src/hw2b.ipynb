### Bullet 3: Experiment with the number of hidden layers

I experimented with a 3 hidden layerd MLP and decieded to go with the settings: 250 neurons/layer, 0.004 learning rate, 0.000005 as regulization term. I decreased the learning reate to be the smallest which still not results in overfit (0.004) and then modified the L! and L2 term aswell with the same reasoning. I then chose the number of neurons to be 250, which results in decent computation time as well as accuracy.

```python
bestScores_1 = []
runTimes_1 = []
for i in range (1,9):
    A, score, time = test_mlp(0, i, 250, 0.004, 0.000005, 0.000005)
    bestScores_1.append(score)
    runTimes_1.append(time)
    print(str(i) + '. iteration complete')
```

```python
A = np.asarray(bestScores_1)
plt.plot(np.arange(1,len(A)+1),1-A)
plt.xlabel('Number of hidden layer')
plt.ylabel('Prediction accuracy')
plt.show
```

As it can be seen, The accuracy varies with the number of hidden layers. The best one was achieved with 8 layers, and it has an increasing tendency, if we not take into accaunt the poor performance with 7 layers, where the model overfits early.
