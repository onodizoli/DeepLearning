#### Bullet 2: Deep MLP

For Deep MLP i was also able to learn both the 8 and 12 bit parity altaugh with far less neurons. In the case of the shallow MLP I used 128 neurons for 8 bit which had 8 x 128 + 128 x 2 = 1280 weight parameters while for the Deep MLP 8bit had 5 x 8 x 8 + 8 x 2 = 336 weight parameters which is far less, yet since the neurons are spread across layers they are more efficient. For the 12 bit case I used relu activation, and the parameters in the shallow MLP where 4096 neurons and 8 x 4096 + 4096 x 2 = 40960 weight parameters, while in the Deep MLP I used 5 layers, 8 neurons each totaling 40 neurons and 5 x 8 x 8 + 8 x 2 = 336 weight parameters which is again far less then the shallow one.
